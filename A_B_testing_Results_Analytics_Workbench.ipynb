# **A/B Testing Report Analytics - standard steps**


---

Analyzing the performance of variation B over A 



To be a Winner in each metric, B needs to reflect a valid and reliable better performance for the traffic in the respective metrics.

We will analyze the report in the following sequence:
*   Step 1: Sample data must correctly represent the population, if there is a sampling error, the report could not be taken into consideration further 

*  Step 2: Filtering out the metrics with the associated confidence level being higher than 95%, only reliable metrics performance will be taken for further evaluation

*  Step 3: Evaluating the B's performance for the filtered metrics


In industry, the set standards for A/B testing: 

*   Sample size (validity) : 5 days, 1-2 weeks
*   Confidence level (reliability) : >95%, 90% (in a few cases)
*   Uplift in the goal conversion rate (capability):  Value greater than the cost it takes to implement and maintenance

## B -15 Off Accessories


```python
df1.head()
```

### Step 1: Validation of Sample data. It must be correctly represent the population.

```python
import numpy as np

#  Analyzing the traffic window
df1['Start Time (UTC)'] = pd.to_datetime(df1["Start Time (UTC)"])
df1['End Time (UTC)'] = pd.to_datetime(df1['End Time (UTC)'])
df1['Start Time (Weekday)'] = df1['Start Time (UTC)'].dt.day_name()
df1['End Time (Weekday)'] = df1['End Time (UTC)'].dt.day_name()
df1['Traffic Duration'] = df1['End Time (UTC)'] - df1['Start Time (UTC)'] 
df1['Traffic Duration'] = df1['Traffic Duration']/ np.timedelta64(1,'D').astype('int')
df1['Start Time (Weekday)'] = df1['Start Time (UTC)'].dt.day_name()
df1['End Time (Weekday)'] = df1['End Time (UTC)'].dt.day_name()

```

```python
from matplotlib import pyplot
df1_sample_size = df1[['Variation Name','Metric Event Name','Start Time (UTC)','Start Time (Weekday)','End Time (Weekday)','End Time (UTC)', 'Traffic Duration']]
df1_sample_size
```

**The traffic sample from Wednesday to Monday can cover the behaviour of the traffic for weekdays and weekend across the year. As per the set standards, the provided traffic window could be choosen for representing the traffic behavior across the year. The sampling complexity here comes with the time dimension. For more data, more time we will loose on testing. So we need to optimize the time window as per various complexities. We can continue with this sample.**

###Step 2: Filtering out metrics with the associated confidence level being greater than 95%

```python
significant_metrics = df1.reset_index(drop = True)[df1['Statistical Significance']>.95]['Metric Event Name'].values
print(significant_metrics)
```

```python
df1_significant_metric = df1[df1['Metric Event Name'] == '"Page Visits on Checkout Page"']
df1_significant_metric['Metric Value Lower Range'] = df1_significant_metric['Metric Value'] - df1_significant_metric['Confidence Interval - Low']
df1_significant_metric['Metric Value Higher Range'] = df1_significant_metric['Metric Value'] + df1_significant_metric['Confidence Interval - High']
```

```python
display(df1_significant_metric[['Variation Name','Metric Event Name','Metric Value','Metric Value Lower Range','Metric Value Higher Range','Statistical Significance']])
```

```python
uplift = (df1_significant_metric['Metric Value'].values[0] - df1_significant_metric['Metric Value'].values[1])/ df1_significant_metric['Metric Value'].values[1] * 100

display("Uplift(%):"+  str(round(uplift,1)))
```

```python

confidence_interval = [round((df1_significant_metric['Metric Value Lower Range'].values[0] - df1_significant_metric['Metric Value'].values[1])/df1_significant_metric['Metric Value'].values[1] * 100),round((df1_significant_metric['Metric Value Higher Range'].values[0] - df1_significant_metric['Metric Value'].values[1])/df1_significant_metric['Metric Value'].values[1] * 100)]
display("Confidence Interval (%): "+str(confidence_interval))
```

With 99% confidence level, there is a uplift of 13.6% in the performance of the metric - "Page Visits on Checkout page". Winner!
